## Mpg

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

### Box plot

```python
sns.boxplot(x = mpg.origin, y = mpg.mpg, 
            order=['japan', 'europe', 'usa'], hue = mpg.cylinders)
plt.show()
```

## Titanic

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

### Scatter plot

```python
sns.scatterplot(x = titanic.age, y = titanic.fare ,data = titanic)
plt.show()
```

```python
sns.scatterplot(x = titanic.age, y = titanic.fare, data = titanic, hue = titanic.sex)
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = titanic.age, y = titanic.fare, data=titanic,
                hue = titanic.sex, size = titanic.who, style = titanic.who,
                sizes=(100,200))
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = titanic.who, y = titanic.fare, data = titanic,
                hue = titanic.alive, size = titanic.who, style = titanic.alive,
                sizes = (100,200), palette = 'flare', alpha = 0.7)
plt.show()
```

### Multi plots

```python
plt.figure(figsize = (16,9))
sns.barplot(x = titanic.age, y = titanic.fare, data = titanic)
sns.lineplot(x = titanic.age, y = titanic.fare, data = titanic)
sns.scatterplot(x = titanic.age, y = titanic.fare, data = titanic)
plt.show()
```
